```python
#Today's Lab Tasks 09/11/2021

#TASK:1

class Node:
  def __init__(self, item = None):
    self.item = item
    self.next = None

class LinkedList:               
  def __init__(self):
    self.head = None
    self.ctr = 0

  def PrintList(self):
      temp = self.head
      elements = []
      print("Linked List contains:", end = " ")
      while temp is not None:
        elements.append(temp.item)
        temp = temp.next
      for i in elements:
        print(i, end = " ")
      print("\n")


  def ReverseList(self):
    temp = self.head
    elements = []

    print("Reversed Linked List:", end = " ")
    while temp is not None:
        elements.append(temp.item)
        temp = temp.next
    for i in reversed(elements): 
        print(i, end = " ")
    print("\n")

  def AppendLeft(self, newItem): 
      newNode = Node(newItem)
      newNode.next = self.head
      self.head = newNode
      self.ctr += 1
      print("New President: ", newItem)
      return

  def AppendEnd(self, newItem):
    newNode = Node(newItem)
    temp = self.head

    if temp is None: 
        self.head = newNode
        return
    while temp.next is not None:
        temp = temp.next
    temp.next = newNode
    self.ctr += 1
    print("New Secretary: ", newItem)
    return
  
  def AppendAt(self, pos, newItem):
      newNode = Node(newItem)
      if pos < 1 or pos > self.ctr:
        print("\tInvalid position!")
        return
      else:
        temp = self.head
        i = 1
      while i < pos:
        temp = temp.next
        i += 1
      newNode.next = temp.next
      temp.next = newNode

      self.ctr += 1
      print("New item Appended:",newItem, "at position:",pos)
      return

  def DeleteLeft(self):     
    if self.head == None:
       print("There are no nodes")
    else: 
      temp = self.head
      second = self.head.next
      
      self.head.next = None     
      self.head = second
      self.ctr -= 1
      print("New president is: ", second.item)

  def DeleteEnd(self):
    temp = self.head
    while(temp.next is not None):
      secondLast = temp
      temp = temp.next
    secondLast.next = None
    self.ctr -= 1
    print("New secretary is: ", secondLast.item)


  def DeleteAt(self, pos):
      if self.head == None:
        print("No nodes exist")
      elif pos == 0:
        self.DeleteLeft()
      elif pos == self.ctr:
       self.DeleteEnd()
      else:
        temp = self.head
        prev = temp
        i = 0
        while (i < int(pos)):
          prev = temp
          temp = temp.next
          i += 1
          prev.next = temp.next
        print("Node deleted: ", temp.item)
        temp.next = None
      self.ctr -= 1


  def GetSize(self): 
    print("Size of Linked list: ", self.ctr)    
    return




def Menu():
    print("1. Add New President")
    print("2. Add at position")
    print("3. Add New Secretary")
    print("4. Delete President")
    print("5. Delete from postion")
    print("6. Delete Secretary")
    print("7. Print Linked List")
    print("8. Print Reversed")
    print("9. Size of Linked List")
    print("10. Exit")
    ch = eval(input("Enter choice: "))
    return ch

print("\t*Welcome to ‘EE Sports Club*\n")
myList = LinkedList()  



while True:
  ch = Menu()
  if ch == 1:
    pName = input("Enter president name: ")
    myList.AppendLeft(pName)
  elif ch == 2:
    name = input("Enter Name of member:")
    pos = int(input("Enter the position:"))
    myList.AppendAt(pos, name)
  elif ch == 3:
    sName = input("Enter name of secretary:")
    myList.AppendEnd(sName)
  elif ch == 4:
    myList.DeleteLeft()
  elif ch == 5:
    pos = input("Enter position:")
    myList.DeleteAt(pos)
  elif ch == 6:
    myList.DeleteEnd()
  elif ch == 7:
    myList.PrintList()
  elif ch == 8:
    myList.ReverseList()
  elif ch == 9:
    myList.GetSize()
  else:
    print("Exit")
    break
print("\n\n")

```
